```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
# imorting libraries
```

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
# Splitting variables
X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

```python
x=df['R&D Spend']
y=df['Profit']
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
# Implementing the model
lin_reg=LinearRegression()
# fit model in training data
lin_reg.fit(X_train, y_train)
```

```python
# y intercept
b0=lin_reg.intercept_
print(b0)

# coefficient
b1=lin_reg.coef_
print(b1)
```

```python
# Prediction result
y_pred_test = lin_reg.predict(X_test)     # predicted value of y_test
y_pred_train = lin_reg.predict(X_train)   # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'green')
plt.title('R&D Spend vs Profit (Training Set)')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'profit', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Prediction on testing set
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'green')
plt.title('R&D Spend vs Profit (Testing Set)')
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'profit', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
